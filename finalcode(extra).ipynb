```python
import torch
import pytesseract
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration, T5Tokenizer, T5ForConditionalGeneration

# Load BLIP-2 model for more accurate image captioning
caption_model = "Salesforce/blip-image-captioning-large"
processor = BlipProcessor.from_pretrained(caption_model)
blip_model = BlipForConditionalGeneration.from_pretrained(caption_model)

# Load Flan-T5 Large for improved question generation
t5_model = "google/flan-t5-large"
tokenizer = T5Tokenizer.from_pretrained(t5_model)
llm_model = T5ForConditionalGeneration.from_pretrained(t5_model)

# Function to generate a highly detailed caption
def generate_caption(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(image, return_tensors="pt")
    output = blip_model.generate(**inputs, max_length=50, num_beams=5)  # More beams = better quality
    caption = processor.decode(output[0], skip_special_tokens=True)
    return caption

# Function to improve the caption for deeper context
def enhance_caption(caption):
    return f"{caption}. Think about its appearance, purpose, and symbolic meaning."

# Function to extract text from an image using OCR
def extract_text_from_image(image_path):
    image = Image.open(image_path)
    extracted_text = pytesseract.image_to_string(image).strip()
    return extracted_text if extracted_text else None

# Function to generate highly accurate questions
def generate_questions(text):
    prompt_text = (
        f"Generate 5 high-quality, context-aware questions based on: {text}. "
        "The questions should explore details, symbolism, scientific properties, and real-world relevance."
    )

    input_ids = tokenizer(prompt_text, return_tensors="pt").input_ids
    output_ids = llm_model.generate(
        input_ids,
        max_length=150,
        num_return_sequences=5,
        num_beams=5,  # Better quality with beam search
        temperature=0.5,  # Lower temp = more deterministic, better accuracy
        top_k=20  # Filters out low-quality questions
    )

    return [tokenizer.decode(output, skip_special_tokens=True) for output in output_ids]

# Function to process an image and generate top-quality questions
def process_image(image_path):
    extracted_text = extract_text_from_image(image_path)

    if extracted_text:
        print("\n📖 **Extracted Text from Image:**\n", extracted_text)
        questions = generate_questions(extracted_text)
    else:
        caption = generate_caption(image_path)
        enhanced_caption = enhance_caption(caption)
        print("\n🖼️ **Generated Caption:**", enhanced_caption)
        questions = generate_questions(enhanced_caption)

    print("\n🎯Generated promts:**")
    for i, question in enumerate(questions, 1):
        print(f"{i}. {question}")

# Example: Upload and process an image
image_path = "/content/poem.jpeg"  # Replace with your uploaded image path
process_image(image_path)
```
